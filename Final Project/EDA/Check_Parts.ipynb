```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import json
```

```python
train_categories = {
    'part':[],
    'category1':[],
    'category2':[],
    'category3':[],
    'category4':[],
    'category5':[],
}

with open("../dataset/train/annotations.json", "r") as json_file:
    train_ann = json.load(json_file)
images = train_ann['images']

for image in images:
    file_name = image['file_name']
    file_path = file_name.replace('jpg', 'json')

    with open(os.path.join("../dataset/train", file_path), "r") as json_file:
        img_json = json.load(json_file)
    
    for key in img_json.keys():
        if key == 'file_name': break
        train_categories[key].append(img_json[key])  
```

```python
valid_categories = {
    'part':[],
    'category1':[],
    'category2':[],
    'category3':[],
    'category4':[],
    'category5':[],
}

with open("../dataset/valid/annotations.json", "r") as json_file:
    valid_ann = json.load(json_file)
images = valid_ann['images']

for image in images:
    file_name = image['file_name']
    file_path = file_name.replace('jpg', 'json')

    with open(os.path.join("../dataset/valid", file_path), "r") as json_file:
        img_json = json.load(json_file)
    
    for key in img_json.keys():
        if key == 'file_name': break
        valid_categories[key].append(img_json[key])    

```

```python
plt.figure(figsize=(20, 10))

for index, key in enumerate(train_categories.keys()):
    df_train = pd.DataFrame(train_categories[key])
    df_valid = pd.DataFrame(valid_categories[key])
    
    train_values = df_train[0].value_counts().sort_index()
    valid_values = df_valid[0].value_counts().sort_index()

    plt.subplot(2, 3, index+1)
    plt.bar(x=train_values.index, height=train_values.values)
    plt.bar(x=valid_values.index, height=valid_values.values)
    plt.xticks(np.arange(-2, 6))

    for index, value in zip(train_values.index, train_values.values):
        plt.text(index, value, value,
        horizontalalignment='center',
        verticalalignment='bottom')

    for index, value in zip(valid_values.index, valid_values.values):
        plt.text(index, value, value,
        horizontalalignment='center',
        verticalalignment='bottom')

    plt.title(key)
plt.show()
```

```python
for i in range(0, 4):
    train_categories = {
        'category1':[],
        'category2':[],
        'category3':[],
        'category4':[],
        'category5':[],
    }

    with open("../dataset/train/annotations.json", "r") as json_file:
        train_ann = json.load(json_file)
    images = train_ann['images']

    for image in images:
        file_name = image['file_name']
        file_path = file_name.replace('jpg', 'json')

        with open(os.path.join("../dataset/train", file_path), "r") as json_file:
            img_json = json.load(json_file)
        if img_json['part'] != i: continue
        
        for key in img_json.keys():
            if key == 'part' or key == 'file_name': continue
            train_categories[key].append(img_json[key])
    
    valid_categories = {
        'category1':[],
        'category2':[],
        'category3':[],
        'category4':[],
        'category5':[],
    }

    with open("../dataset/valid/annotations.json", "r") as json_file:
        valid_ann = json.load(json_file)
    images = valid_ann['images']

    for image in images:
        file_name = image['file_name']
        file_path = file_name.replace('jpg', 'json')

        with open(os.path.join("../dataset/valid", file_path), "r") as json_file:
            img_json = json.load(json_file)
        if img_json['part'] != i: continue

        for key in img_json.keys():
            if key == 'part' or key == 'file_name': continue
            valid_categories[key].append(img_json[key])  

    plt.figure(figsize=(20, 10))

    for index, key in enumerate(train_categories.keys()):
        df_train = pd.DataFrame(train_categories[key])
        df_valid = pd.DataFrame(valid_categories[key])
        
        train_values = df_train[0].value_counts().sort_index()
        valid_values = df_valid[0].value_counts().sort_index()

        plt.subplot(2, 3, index+1)
        plt.bar(x=train_values.index, height=train_values.values)
        plt.bar(x=valid_values.index, height=valid_values.values)
        plt.xticks(np.arange(-2, 6))

        for index, value in zip(train_values.index, train_values.values):
            plt.text(index, value, value,
            horizontalalignment='center',
            verticalalignment='bottom')

        for index, value in zip(valid_values.index, valid_values.values):
            plt.text(index, value, value,
            horizontalalignment='center',
            verticalalignment='bottom')

        plt.title(f'part:{i} / '+key)
    plt.show()
    
```

```python
index = 0
index_json = []

with open("../dataset/train/annotations.json", "r") as json_file:
    train_ann = json.load(json_file)
images = train_ann['images']

for image in images:
    file_name = image['file_name']
    file_path = file_name.replace('jpg', 'json')

    with open(os.path.join("../dataset/train", file_path), "r") as json_file:
        img_json = json.load(json_file)
    
    if img_json['part']==index: index_json.append(img_json)

print(len(index_json))
with open(f'../dataset/train/part_{index}.json', 'w') as json_file:
    json.dump(index_json, json_file, indent=4)
```
